**Univariate Analysis**

```python
plt.hist(students_df_2['math score'])
# Insight : 60 to 70 score is the highest frequency score in Maths subject.
# Insight : 0 to 20 marks represent lowest frequency score in Maths subject.
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(students_df_2['math score'], kde=True, bins=20) # kde [Kernel Density Estimation] adds a density curve, bins control number of bars
plt.title('Distribution of Math Scores', fontsize=16)
plt.xlabel('Math Score', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.hist(students_df_2['reading score'])
# Insight : Between 60 to 80 marks there exist highest frequency in reading score.
# Insight : 4 to 22 marks represent lowest frequency score in Reading subject.
```

```python
plt.figure(figsize = (8, 6))
sns.histplot(students_df_2['reading score'], kde=True, bins=20) # kde adds a density curve, bins control number of bars
plt.title('Distribution of Reading Scores', fontsize=16)
plt.xlabel('reading score', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.hist(students_df_2['writing score'])
# Insight : Between 60 to 80 marks there exist highest frequency in writing score.
# Insight : 2 to 20 marks represent lowest frequency score in writing subject.
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(students_df_2['writing score'], kde=True, bins=20) # kde adds a density curve, bins control number of bars
plt.title('Distribution of Writing Scores', fontsize=16)
plt.xlabel('writing score', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

**Bivariate Analysis**

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(x='math score', y='reading score', data = students_df_2, palette='coolwarm', s=70, alpha=0.8)
plt.title('Math Score vs. Reading Score', fontsize=16)
plt.xlabel('Math Score', fontsize=14)
plt.ylabel('Reading Score', fontsize=14)
plt.legend(title='Test Preparation')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
# Below graph depicts the dependency of math score and reading score on each other.
```

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(x='math score', y='writing score', data=students_df_2, palette='coolwarm', s=70, alpha=0.8)
plt.title('Math Score vs. Writing Score', fontsize=16)
plt.xlabel('Math Score', fontsize=14)
plt.ylabel('Writing Score', fontsize=14)
plt.legend(title='Test Preparation')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
# Below graph depicts the dependency of math score and writing score on each other.
```

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(x='reading score', y='writing score', data=students_df_2, palette='coolwarm', s=70, alpha=0.8)
plt.title('Reading Score Vs Writting score', fontsize=16)
plt.xlabel('Reading Score', fontsize=14)
plt.ylabel('Writing Score', fontsize=14)
plt.legend(title='Test Preparation')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
# Below graph depicts the dependency of reading score and writing score on each other.
```
