```python
def treat_missing_values( df, num_method = 'median', cat_method = 'mode') :
    for column in df_North_South_1.columns :
        if df_North_South_1[column].isnull().sum() > 0 :

            if df_North_South_1[column].dtype in ['int64', 'float64'] :

                if num_method == 'mean' :
                    df_North_South_1[column].fillna(round(df_North_South_1[column].mean()), inplace = True)
                if num_method == 'median' :
                    df_North_South_1[column].fillna(df_North_South_1[column].median(), inplace = True)
                if num_method == 'mode' :
                    df_North_South_1[column].fillna(df_North_South_1[column].mode()[0], inplace = True)
                else :
                    print("Pass the correct option")

            else :
                if cat_method == 'mode' :
                    df_North_South_1[column].fillna(df_North_South_1[column].mode()[0], inplace = True)
                else :
                    print("Pass the correct option")

    return df_North_South_1
```

#**STEP - 4 : Outlier Treatment**

```python
# Store all categorical columns in seperate list
categorical_columns = list(df_North_South_1.columns[df_North_South_1.dtypes == 'object'])
# Store all numerical columns in seperate list
numerical_columns = list(df_North_South_1.columns[df_North_South_1.dtypes != 'object'])
```

```python
# Now let's Perform "Capping" & "Flooring" for all the columns

for col in numerical_columns :

  q1 = df_North_South_1[col].describe() ["25%"]
  q3 = df_North_South_1[col].describe() ["75%"]
  iqr = q3 - q1
  lower_whisker = q1 - 1.5 * iqr
  upper_whisker = q3 + 1.5 * iqr

  df_North_South_1[col] = np.where(df_North_South_1[col] < lower_whisker, lower_whisker, df_North_South_1[col])
  df_North_South_1[col] = np.where(df_North_South_1[col] > upper_whisker, upper_whisker, df_North_South_1[col])
```
